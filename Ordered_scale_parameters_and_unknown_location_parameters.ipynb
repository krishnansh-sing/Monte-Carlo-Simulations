```python
def U(x,mu1,mu2,sigma1,sigma2):
        dis =  (x-mu1)/sigma1 - (x-mu2)/sigma2 - np.log(sigma2/sigma1) 
        return dis
```

```python
def umR1(rho,delta,n1,n2):

    np.random.seed(42)
    
    # Simulation settings
    num_simulations = 10000

    # Storage for probability results
    term1, term2 = 0, 0

    for _ in range(num_simulations):
        # Generate random integer for sigma2
        sigma2 = random.randint(2, 100)
        # Calculate sigma1 to ensure the ratio sigma1/sigma2 = rho
        sigma1 = sigma2 * rho
        mu1 = random.randint(2, 100)
        mu2 = (delta*sigma2 + mu1)
        sample_P1 = mu1 + np.random.exponential(scale=sigma1, size=n1)
        sample_P2 = mu2 + np.random.exponential(scale=sigma2, size=n2)
        emu1 = min(sample_P1)
        emu2 = min(sample_P2)
        T1 = sum(sample_P1) - n1*emu1
        T2 = sum(sample_P2) - n2*emu2
        est1 = T1/n1
        est2 = T2/n2
        obs1 =  mu1 + np.random.exponential(scale=sigma1, size=1)
        obs2 =  mu2 + np.random.exponential(scale=sigma2, size=1)
        if emu1 <= obs1 < emu2 or ( U(obs1,emu1,emu2,est1,est2) <= 0 and obs1 >= max(emu1,emu2) ):
            term1 = term1 + 1
        if emu2 <= obs2 < emu1 or ( U(obs2,emu1,emu2,est1,est2) > 0 and obs2 >= max(emu1,emu2) ) :
            term2 = term2 + 1
    # Calculate probabilities
    P1 = term1 / num_simulations
    P2 = term2 / num_simulations
    
    # Compute EPC
    EPC =  0.5 * (P1 + P2)
    
    return EPC
```

```python
def umR2(rho,delta,n1,n2,a=0):

    np.random.seed(42)
    
    # Simulation settings
    num_simulations = 10000

    # Storage for probability results
    term1, term2 = 0, 0

    for _ in range(num_simulations):
        # Generate random integer for sigma2
        sigma2 = random.randint(2, 100)
        # Calculate sigma1 to ensure the ratio sigma1/sigma2 = rho
        sigma1 = sigma2 * rho
        mu1 = random.randint(2, 100)
        mu2 = (delta*sigma2 + mu1)
        sample_P1 = mu1 + np.random.exponential(scale=sigma1, size=n1)
        sample_P2 = mu2 + np.random.exponential(scale=sigma2, size=n2)
        emu1 = min(sample_P1)
        emu2 = min(sample_P2)
        T1 = sum(sample_P1) - n1*emu1
        T2 = sum(sample_P2) - n2*emu2
        est1 = min(T1/n1,a*(T1/n1)+(1-a)*(T2/n2))
        est2 = max(T2/n2,a*(T2/n2)+(1-a)*(T1/n1))
        obs1 =  mu1 + np.random.exponential(scale=sigma1, size=1)
        obs2 =  mu2 + np.random.exponential(scale=sigma2, size=1)
        if emu1 <= obs1 < emu2 or ( U(obs1,emu1,emu2,est1,est2) <= 0 and obs1 >= max(emu1,emu2) ):
            term1 = term1 + 1
        if emu2 <= obs2 < emu1 or ( U(obs2,emu1,emu2,est1,est2) > 0 and obs2 >= max(emu1,emu2) ) :
            term2 = term2 + 1
    # Calculate probabilities
    P1 = term1 / num_simulations
    P2 = term2 / num_simulations
    
    # Compute EPC
    EPC =  0.5 * (P1 + P2)
    
    return EPC
```

```python
def umR4(rho,delta,n1,n2):

    np.random.seed(42)
    
    # Simulation settings
    num_simulations = 10000

    # Storage for probability results
    term1, term2 = 0, 0

    for _ in range(num_simulations):
        # Generate random integer for sigma2
        sigma2 = random.randint(2, 100)
        # Calculate sigma1 to ensure the ratio sigma1/sigma2 = rho
        sigma1 = sigma2 * rho
        mu1 = random.randint(2, 100)
        mu2 = (delta*sigma2 + mu1)
        sample_P1 = mu1 + np.random.exponential(scale=sigma1, size=n1)
        sample_P2 = mu2 + np.random.exponential(scale=sigma2, size=n2)
        emu1 = min(sample_P1)
        emu2 = min(sample_P2)
        T1 = sum(sample_P1) - n1*emu1
        T2 = sum(sample_P2) - n2*emu2

        if n2*T1 <= n1*T2 :
            est1 = T1/n1
        elif n1*T2 < n2*T1 <= n2*(n1 + n2 - 1)*T2 :
            est1 = (T1+T2)/(n1+n2)
        elif T1 > (n1 + n2 - 1)*T2 :
            est1 = (T1)/(n1 + n2 - 1)

        if n2*T1 <= (n1 - 1)*T2 :
            est2 = T2/n2
        else :
            est2 = (T1+T2)/(n1+n2-1)
        
        obs1 =  mu1 + np.random.exponential(scale=sigma1, size=1)
        obs2 =  mu2 + np.random.exponential(scale=sigma2, size=1)
        if emu1 <= obs1 < emu2 or ( U(obs1,emu1,emu2,est1,est2) <= 0 and obs1 >= max(emu1,emu2) ):
            term1 = term1 + 1
        if emu2 <= obs2 < emu1 or ( U(obs2,emu1,emu2,est1,est2) > 0 and obs2 >= max(emu1,emu2) ) :
            term2 = term2 + 1
    # Calculate probabilities
    P1 = term1 / num_simulations
    P2 = term2 / num_simulations
    
    # Compute EPC
    EPC =  0.5 * (P1 + P2)
    
    return EPC
```

```python
def U(x,mu1,mu2,sigma1,sigma2):
        dis =  (x-mu1)/sigma1 - (x-mu2)/sigma2 - np.log(sigma2/sigma1) 
        return dis
```

```python
def kmR1(rho,delta,n1,n2):

    np.random.seed(1234)
    
    # Simulation settings
    num_simulations = 10000

    # Storage for probability results
    term1, term2 = 0, 0

    for _ in range(num_simulations):
        # Generate random integer for sigma2
        sigma2 = random.randint(2, 100)
        # Calculate sigma1 to ensure the ratio sigma1/sigma2 = rho
        sigma1 = sigma2 * rho
        mu1 = random.randint(2, 100)
        mu2 = (delta*sigma2 + mu1)
        sample_P1 = mu1 + np.random.exponential(scale=sigma1, size=n1)
        sample_P2 = mu2 + np.random.exponential(scale=sigma2, size=n2)
        est1 = np.mean(sample_P1) - mu1
        est2 = np.mean(sample_P2) - mu2
        obs1 = mu1 + np.random.exponential(scale=sigma1, size=1)
        obs2 = mu2 + np.random.exponential(scale=sigma2, size=1)
        if mu1 <= obs1 < mu2 or ( U(obs1,mu1,mu2,est1,est2) <= 0 and obs1 >= max(mu1,mu2) ):
            term1 = term1 + 1
        if mu2 <= obs2 < mu1 or ( U(obs2,mu1,mu2,est1,est2) > 0 and obs2 >= max(mu1,mu2) ) :
            term2 = term2 + 1
    # Calculate probabilities
    P1 = term1 / num_simulations
    P2 = term2 / num_simulations
    
    # Compute EPC
    EPC =  0.5 * (P1 + P2)
    
    return EPC
```

```python
def kmR2(rho,delta,n1,n2,a=0):
    
    np.random.seed(1234)
    
    # Simulation settings
    num_simulations = 10000

    # Storage for probability results
    term1, term2 = 0, 0

    for _ in range(num_simulations):
        # Generate random integer for sigma2
        sigma2 = random.randint(2, 100)
        # Calculate sigma1 to ensure the ratio sigma1/sigma2 = rho
        sigma1 = sigma2 * rho
        mu1 = random.randint(2, 100)
        mu2 = (delta*sigma2 + mu1)
        sample_P1 = mu1 + np.random.exponential(scale=sigma1, size=n1)
        sample_P2 = mu2 + np.random.exponential(scale=sigma2, size=n2)
        est1 = min(np.mean(sample_P1) - mu1,a*(np.mean(sample_P1) - mu1)+(1-a)*(np.mean(sample_P2) - mu2))
        est2 = max(np.mean(sample_P2) - mu2,a*(np.mean(sample_P2) - mu2)+(1-a)*(np.mean(sample_P1) - mu1))
        obs1 = mu1 + np.random.exponential(scale=sigma1, size=1)
        obs2 = mu2 + np.random.exponential(scale=sigma2, size=1)
        if mu1 <= obs1 < mu2 or ( U(obs1,mu1,mu2,est1,est2) <= 0 and obs1 >= max(mu1,mu2) ):
            term1 = term1 + 1
        if mu2 <= obs2 < mu1 or ( U(obs2,mu1,mu2,est1,est2) > 0 and obs2 >= max(mu1,mu2) ) :
            term2 = term2 + 1
    # Calculate probabilities
    P1 = term1 / num_simulations
    P2 = term2 / num_simulations
    
    # Compute EPC
    EPC =  0.5 * (P1 + P2)
    
    return EPC
```

```python
def kmR3(rho,delta,n1,n2):
    
    np.random.seed(1234)
    
    # Simulation settings
    num_simulations = 10000

    # Storage for probability results
    term1, term2 = 0, 0

    for _ in range(num_simulations):
        # Generate random integer for sigma2
        sigma2 = random.randint(2, 100)
        # Calculate sigma1 to ensure the ratio sigma1/sigma2 = rho
        sigma1 = sigma2 * rho
        mu1 = random.randint(2, 100)
        mu2 = (delta*sigma2 + mu1)
        sample_P1 = mu1 + np.random.exponential(scale=sigma1, size=n1)
        sample_P2 = mu2 + np.random.exponential(scale=sigma2, size=n2)
        est1 = min(np.mean(sample_P1) - mu1,(n1*(np.mean(sample_P1) - mu1) + n2*(np.mean(sample_P2) - mu2))/(n1 +n2))
        est2 = max(np.mean(sample_P2) - mu2,(n1*(np.mean(sample_P1) - mu1) + n2*(np.mean(sample_P2) - mu2))/(n1 +n2))
        obs1 = mu1 + np.random.exponential(scale=sigma1, size=1)
        obs2 = mu2 + np.random.exponential(scale=sigma2, size=1)
        if mu1 <= obs1 < mu2 or ( U(obs1,mu1,mu2,est1,est2) <= 0 and obs1 >= max(mu1,mu2) ):
            term1 = term1 + 1
        if mu2 <= obs2 < mu1 or ( U(obs2,mu1,mu2,est1,est2) > 0 and obs2 >= max(mu1,mu2) ) :
            term2 = term2 + 1
    # Calculate probabilities
    P1 = term1 / num_simulations
    P2 = term2 / num_simulations
    
    # Compute EPC
    EPC =  0.5 * (P1 + P2)
    
    return EPC
```

```python
def kmR4(rho,delta,n1,n2):
    
    np.random.seed(1234)
    
    # Simulation settings
    num_simulations = 10000

    # Storage for probability results
    term1, term2 = 0, 0

    for _ in range(num_simulations):
        # Generate random integer for sigma2
        sigma2 = random.randint(2, 100)
        # Calculate sigma1 to ensure the ratio sigma1/sigma2 = rho
        sigma1 = sigma2 * rho
        mu1 = random.randint(2, 100)
        mu2 = (delta*sigma2 + mu1)
        sample_P1 = mu1 + np.random.exponential(scale=sigma1, size=n1)
        sample_P2 = mu2 + np.random.exponential(scale=sigma2, size=n2)

        if (n2 + 1)*(np.mean(sample_P1) - mu1) <= n2*(np.mean(sample_P2) - mu2):
            est1 = np.mean(sample_P1)-mu1
        else:
            est1 = (n1*(np.mean(sample_P1) - mu1) + n2*(np.mean(sample_P2) - mu2))/(n1+n2+1)
            
        est2 = max((n1*(np.mean(sample_P1) - mu1) + n2*(np.mean(sample_P2) - mu2))/(n1 +n2),np.mean(sample_P2) - mu2)
            
        obs1 = mu1 + np.random.exponential(scale=sigma1, size=1)
        obs2 = mu2 + np.random.exponential(scale=sigma2, size=1)
        if mu1 <= obs1 < mu2 or ( U(obs1,mu1,mu2,est1,est2) <= 0 and obs1 >= max(mu1,mu2) ):
            term1 = term1 + 1
        if mu2 <= obs2 < mu1 or ( U(obs2,mu1,mu2,est1,est2) > 0 and obs2 >= max(mu1,mu2) ) :
            term2 = term2 + 1
    # Calculate probabilities
    P1 = term1 / num_simulations
    P2 = term2 / num_simulations
    
    # Compute EPC
    EPC =  0.5 * (P1 + P2)
    
    return EPC
```

```python
def kmR5(rho,delta,n1,n2):
    
    np.random.seed(1234)
    
    # Simulation settings
    num_simulations = 10000

    # Storage for probability results
    term1, term2 = 0, 0

    for _ in range(num_simulations):
        # Generate random integer for sigma2
        sigma2 = random.randint(2, 100)
        # Calculate sigma1 to ensure the ratio sigma1/sigma2 = rho
        sigma1 = sigma2 * rho
        mu1 = random.randint(2, 100)
        mu2 = (delta*sigma2 + mu1)
        sample_P1 = mu1 + np.random.exponential(scale=sigma1, size=n1)
        sample_P2 = mu2 + np.random.exponential(scale=sigma2, size=n2)

        if n1*(np.mean(sample_P1) - mu1) <= (n1+1)*(np.mean(sample_P2) - mu2):
            est1 = n1*(np.mean(sample_P1)-mu1)/(n1 + 1)
        else:
            est1 = (n1*(np.mean(sample_P1) - mu1) + n2*(np.mean(sample_P2) - mu2))/(n1 +n2+1)
            
        
        if (n2 + 1)*(np.mean(sample_P1) - mu1) <= n2*(np.mean(sample_P2) - mu2):
            est2 = n2*(np.mean(sample_P2)-mu2)/(n2 + 1)
        else:
            est2 = (n1*(np.mean(sample_P1) - mu1) + n2*(np.mean(sample_P2) - mu2))/(n1 +n2+1)
            
        obs1 = mu1 + np.random.exponential(scale=sigma1, size=1)
        obs2 = mu2 + np.random.exponential(scale=sigma2, size=1)
        if mu1 <= obs1 < mu2 or ( U(obs1,mu1,mu2,est1,est2) <= 0 and obs1 >= max(mu1,mu2) ):
            term1 = term1 + 1
        if mu2 <= obs2 < mu1 or ( U(obs2,mu1,mu2,est1,est2) > 0 and obs2 >= max(mu1,mu2) ) :
            term2 = term2 + 1
    # Calculate probabilities
    P1 = term1 / num_simulations
    P2 = term2 / num_simulations
    
    # Compute EPC
    EPC =  0.5 * (P1 + P2)
    
    return EPC
```
